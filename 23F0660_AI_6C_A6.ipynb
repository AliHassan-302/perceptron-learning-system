```python
### import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def activation(self, x):
        return 1 if x >= 0 else 0

    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # initialize weights & bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        # training loop
        for epoch in range(self.epochs):
            print(f"\nEpoch {epoch+1}")
            for i in range(n_samples):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = self.activation(linear_output)

                # update rule
                update = self.lr * (y[i] - y_pred)
                self.weights += update * X[i]
                self.bias += update

                print(f"Sample {i+1}: Pred={y_pred}, Actual={y[i]}, Update={update}")

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return [self.activation(x) for x in linear_output]


# ---------------- USER INPUT ----------------

# number of samples & features
n_samples = int(input("Enter number of samples: "))
n_features = int(input("Enter number of features: "))

X = []
y = []

print("\nEnter feature values:")
for i in range(n_samples):
    row = list(map(float, input(f"Sample {i+1} features: ").split()))
    X.append(row)

print("\nEnter labels (0 or 1):")
for i in range(n_samples):
    label = int(input(f"Label for sample {i+1}: "))
    y.append(label)

X = np.array(X)
y = np.array(y)

# hyperparameters
lr = float(input("\nEnter learning rate: "))
epochs = int(input("Enter number of epochs: "))

# train model
model = Perceptron(learning_rate=lr, epochs=epochs)
model.fit(X, y)

print("\nFinal Weights:", model.weights)
print("Final Bias:", model.bias)

# ---------------- PREDICTION ----------------

test_samples = int(input("\nEnter number of test samples: "))

X_test = []
for i in range(test_samples):
    row = list(map(float, input(f"Test sample {i+1}: ").split()))
    X_test.append(row)

X_test = np.array(X_test)

predictions = model.predict(X_test)

print("\nPredictions:")
for i, pred in enumerate(predictions):
    print(f"Sample {i+1}: {pred}")
```
